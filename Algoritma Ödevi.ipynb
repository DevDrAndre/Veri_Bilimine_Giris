```python
#Soru-1

def en_buyuk_sayi(liste):
    en_buyuk = liste[0]
    
    for sayi in liste:
        if sayi > en_buyuk:
            en_buyuk = sayi
    
    return en_buyuk
    
print(en_buyuk_sayi([6, 7, 8, 9, 3]))
```

```python
# Soru-2 

def ters(metin):
    return metin[::-1]

print(ters("Hello World!")) 
print(ters("Python"))
```

```python
#Soru-3

def faktoriyel(n):
    sonuc = 1
    for i in range(1, n + 1):
        sonuc *= i
    return sonuc

print(faktoriyel(7))
```

```python
#Soru-4

def cift(liste):
    return [sayi for sayi in liste if sayi % 2 == 0]

print(cift([1, 2, 3, 4, 5, 6]))
print(cift([11, 23, 44]))
```

```python
#Soru-5

def tek(l1, l2):
    return list(set(l1 + l2))

print(tek([1, 2, 3], [3, 4, 5]))
print(tek([], [1, 1, 2]))
```

```python
#Soru-6

def palindrom_mu(metin):
    temiz = ''.join(metin.lower().split())
    return temiz == temiz[::-1]

print(palindrom_mu("A man a plan a canal Panama"))
print(palindrom_mu("hello"))
```

```python
#Soru-7

def ebob(a, b):
    while b:
        a, b = b, a % b
    return a

print(ebob(54, 24))
print(ebob(7, 13))
```

```python
#Soru-8

def anagram_mi(s1, s2):
    temiz1 = sorted(s1.replace(" ", "").lower())
    temiz2 = sorted(s2.replace(" ", "").lower())
    return temiz1 == temiz2

print(anagram_mi("listen", "silent"))
print(anagram_mi("hello", "world"))
```

```python
#Soru-9

def eksik_sayi(liste):
    n = len(liste) + 1
    toplam = n * (n + 1) // 2
    return toplam - sum(liste)

print(eksik_sayi([1, 2, 4, 5]))
print(eksik_sayi([9, 6, 4, 2, 3, 5, 7, 0, 1]))
```

```python
#Soru-10

def liste_dondur(liste, k):
    k = k % len(liste)
    return liste[-k:] + liste[:-k]

print(liste_dondur([1, 2, 3, 4, 5], 2))
print(liste_dondur([-1, -100, 3, 99], 2))
```

```python
#Soru-11

def median_liste(nums1, nums2):
    birlesik = sorted(nums1 + nums2)
    n = len(birlesik)
    ort = n // 2
    if n % 2 == 0:
        return (birlesik[ort - 1] + birlesik[ort]) / 2
    else:
        return float(birlesik[ort])

print(median_liste([1, 3], [2]))
print(median_liste([1, 2], [3, 4]))
```

```python
#Soru-12

def en_uzun_palindrom(s):
    en_uzun = ""
    for i in range(len(s)):
        for j in range(i, len(s)):
            alt = s[i:j+1]
            if alt == alt[::-1] and len(alt) > len(en_uzun):
                en_uzun = alt
    return en_uzun

print(en_uzun_palindrom("babad"))
print(en_uzun_palindrom("cbbd"))
```

```python
#Soru-14

def max_alt_dizi_toplami(nums):
    max_toplam = nums[0]
    mevcut_toplam = 0
    for n in nums:
        mevcut_toplam = max(n, mevcut_toplam + n)
        max_toplam = max(max_toplam, mevcut_toplam)
    return max_toplam

print(max_alt_dizi_toplami([-2,1,-3,4,-1,2,1,-5,4]))
print(max_alt_dizi_toplami([5,4,-1,7,8]))
```
